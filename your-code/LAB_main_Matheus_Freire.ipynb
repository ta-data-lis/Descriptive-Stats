## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
dice = [1, 2, 3, 4, 5, 6]

def rolling_dice(n_rolls):
    rolls = random.choices(dice, k = n_rolls)
    data = {"Roll": range(1, n_rolls + 1), "Result": rolls}
    df = pd.DataFrame(data)
    return df

dice_df = rolling_dice(10)
dice_df
```

#### 2.- Plot the results sorted by value.

```python
dice_df = dice_df.sort_values("Result")
y = dice_df[["Roll"]]
x = dice_df["Result"]

plt.plot(x,y)
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
freq_table = dice_df.groupby("Result").size().reset_index(name="Frequency")

plt.bar(freq_table["Result"], freq_table["Frequency"])
plt.xlabel("Result")
plt.ylabel("Frequency")
plt.title("Frequency Distribution of Rolling a 10-Sided Dice")
plt.show()
```

```python
"""
Both provide information about the distribution of the result values when rolling a 10-sided die
"""
```
