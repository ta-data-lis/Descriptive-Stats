A Cartesian Product of two sets A and B, written in A x B, is the set containing ordered pairs
from A and B. That is, if C = A x B, then each element of C is of the form (x,y), where:
x ∈ A and y ∈ B.

A x B = {(x,y) | x ∈ A and y ∈ B}

Inclusion-Exclusion principle: |A U B| = |A| + |B| - |A ∩ B|

DISJOINT VS INDEPENDANCE

Data Squad | Bayesian Probability

Disjoint:
A and B cannot occur at the same time e.g. night and day. It looks like 2 circles that have no intersection. it is dependant though. No information in A means it is in B
- A ∩ B = 0
- P (A U B) = P(A) + P(B)

Independant:
A does not give any information on B
- P(A|B) = P(A), P(B|A) = P(B)
- P(A ∩ B) = P(A)P(B)
